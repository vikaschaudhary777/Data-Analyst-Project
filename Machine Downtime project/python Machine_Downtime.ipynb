## Measures of Dispersion

```python
import statsmodels.api as sm

# Select the columns for dispersion calculation
columns_to_analyze = ['Hydraulic_Pressure', 'Coolant_Pressure', 'Air_System_Pressure', 'Coolant_Temperature', 'Hydraulic_Oil_Temperature', 'Spindle_Bearing_Temperature', 'Spindle_Vibration', 'Tool_Vibration', 'Spindle_Speed', 'Voltage', 'Torque', 'Cutting']

# Calculate measures of dispersion for the columns
dispersion_measures = pd.DataFrame({
    'min': mach_down[columns_to_analyze].min(),
    'max': mach_down[columns_to_analyze].max(),
    'std': mach_down[columns_to_analyze].std(),
    'var': mach_down[columns_to_analyze].var(),
    'mad': sm.robust.scale.mad(mach_down[columns_to_analyze]),
    'range': mach_down[columns_to_analyze].max() - mach_down[columns_to_analyze].min()
})

# Print the dispersion measures
print(dispersion_measures)
```

## Outlier Detection

```python
# Set the style of the plot
sns.set(style="whitegrid")

# Select the columns for outlier detection
columns_to_analyze = ['Hydraulic_Pressure', 'Coolant_Pressure', 'Air_System_Pressure',
                      'Coolant_Temperature', 'Hydraulic_Oil_Temperature', 'Spindle_Bearing_Temperature',
                      'Spindle_Vibration', 'Tool_Vibration', 'Spindle_Speed', 'Voltage', 'Torque', 'Cutting']

# Create a color palette
palette = sns.color_palette("husl", len(columns_to_analyze))

# Create box plots for outlier detection
plt.figure(figsize=(14, 8))
box = sns.boxplot(data=mach_down[columns_to_analyze], palette=palette)

# Rotate x-axis labels for better readability
box.set_xticklabels(box.get_xticklabels(), rotation=45, horizontalalignment='right')

# Add a title and labels
plt.title('Box Plots - Outlier Detection', fontsize=16)
plt.xlabel('Parameters', fontsize=14)
plt.ylabel('Values', fontsize=14)

# Add gridlines for better readability
plt.grid(True, linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```

## scatter plots for outlier detection

```python
# Set the matplotlib backend to display inline
%matplotlib inline

# Set the style of the plot
sns.set(style="whitegrid")

# Select the columns for outlier detection
columns_to_analyze = ['Hydraulic_Pressure', 'Coolant_Pressure', 'Air_System_Pressure',
                      'Coolant_Temperature', 'Hydraulic_Oil_Temperature', 'Spindle_Bearing_Temperature',
                      'Spindle_Vibration', 'Tool_Vibration', 'Spindle_Speed', 'Voltage', 'Torque', 'Cutting']

# Create scatter plots for outlier detection
for column in columns_to_analyze:
    plt.figure(figsize=(10, 6))
    plt.scatter(mach_down[column], mach_down[column], c='blue', label='Data', alpha=0.6)

    # Identify outliers and change their color to red
    outliers = mach_down[column].loc[mach_down[column] > mach_down[column].mean() + 2 * mach_down[column].std()]
    plt.scatter(outliers, outliers, c='red', label='Outliers', alpha=0.8)

    plt.xlabel(column, fontsize=12)
    plt.ylabel('Values', fontsize=12)
    plt.title(f'Scatter Plot - {column} (Outlier Detection)', fontsize=14)
    plt.legend(loc='upper right')

    # Add gridlines for better readability
    plt.grid(True, linestyle='--', alpha=0.7)

    # Show the plot
    plt.tight_layout()
    plt.show()
```

```python
# Convert 'Date' column to datetime format
machine_downtime_counts.loc[:, 'Date'] = pd.to_datetime(machine_downtime_counts['Date'], dayfirst=True, format='%d/%m/%Y', errors='coerce')

# Filter the dataframe and count occurrences of 'machine_failure'
mdc = machine_downtime_counts[machine_downtime_counts['Downtime'] == 'Machine_Failure'].groupby(pd.Grouper(freq='W', key='Date')).count()

# Create the plot
plt.figure(figsize=(20, 10))

# Plot the distribution with markers and line
plt.plot(mdc.index, mdc.values, marker='o', linestyle='-', color='b', markersize=8, linewidth=2, label='Downtime Occurrences')

# Add grid lines for better readability
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# Enhance the plot with labels and title
plt.xlabel('Week', fontsize=14)
plt.ylabel('Downtime Occurrences', fontsize=14)
plt.title('Distribution of Downtime Occurrences over Time (Grouped by Week)', fontsize=16)
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)

# Add a legend to the plot
plt.legend()

# Show the plot
plt.tight_layout()  # Adjust layout to fit labels
plt.show()
```

```python
# Plot the histogram by date
sns.histplot(data=mdc, x ='Date', bins=10)
plt.xlabel('Date')
plt.ylabel('Frequency')
plt.title('Histogram of Downtime Occurrences by Date')
plt.xticks(rotation=45)
plt.show()
```

```python
# Reset the index to prepare for plotting
mdc_mid = mdc_mid.reset_index()

# Rename the columns for clarity
mdc_mid.columns = ['Machine_ID', 'Failure_Count']

# Function to display percentage and count on the pie chart
def autopct_format(values):
    def my_format(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return f'{pct:.1f}%\n({val})'
    return my_format

# Plot the pie chart
plt.figure(figsize=(7, 7))
plt.pie(mdc_mid['Failure_Count'], labels=mdc_mid['Machine_ID'], autopct='%1.1f%%', startangle=90, colors=sns.color_palette('viridis', len(mdc_mid)))

# Enhance the plot with a title
plt.title('Machine Failures per Machine ID', fontsize=16)

# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')

# Show the plot
plt.show()
```
